```python
# importing the required computer vision library

import cv2
```

```python
# In order to show the image, we use imshow() function

from google.colab.patches import cv2_imshow
```

```python
"""CASCADE CLASSIFIER: 
Cascading classifiers are trained with several hundred "positive" sample views of a 
particular object and arbitrary "negative" images of the same size. After the classifier
is trained it can be applied to a region of an image and detect the object in question."""

# cvtColor: use to change color either from RGB to B&W or vice versa

# detectMultiScale is used to detect objects of different sizes in the image 
# [[x-coordinate, y-coordinate, width, height]]
```

```python
def face_detect(img_path):
  face_cascade = cv2.CascadeClassifier("/content/haarcascade_frontalface_default.xml")
  
  img = cv2.imread(img_path)
  
  gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
  
  faces = face_cascade.detectMultiScale(gray)
  
  for (x, y, w, h) in faces:
    cv2.rectangle(img, (x, y), (x+w, y+h), (200, 250, 0), 2)
  
  cv2_imshow(img)
```

```python
def eye_detect(img_path):
  eye_cascade = cv2.CascadeClassifier("/content/haarcascade_eye.xml")

  img = cv2.imread(img_path)

  gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)

  eyes = eye_cascade.detectMultiScale(gray)

  for (x, y, w, h) in eyes:
    cv2.rectangle(img, (x,y), (x+w, y+h), (100, 10, 250), 2)

  cv2_imshow(img)
```

```python
img_path = "/content/5d020eb457ae2.jpg"
```

```python
# face detection example 1

face_detect("/content/human pic.webp")
```

```python
# face detection example 2

face_detect(img_path)
```

```python
# eyes detection example 1

eye_detect("/content/human pic.webp")
```

```python
# eyes detection example 2

eye_detect(img_path)
```
